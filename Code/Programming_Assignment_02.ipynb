```python
# Import all the libraries needed
import pandas as pd # to load dataset
import numpy as np # for mathematic equation
from nltk.corpus import stopwords # to get collection of stopwords
from tensorflow.keras.preprocessing.text import Tokenizer # to encode text to int
from tensorflow.keras.preprocessing.sequence import pad_sequences # to do padding or truncating
from tensorflow.keras.models import Sequential # the model
from tensorflow.keras.layers import Embedding, LSTM, Bidirectional, Dense # layers of the architecture
from tensorflow.keras.callbacks import ModelCheckpoint # save model
import re
import nltk
import matplotlib.pyplot as plt
nltk.download('stopwords')

# Declaring the english stop words
english_stops = set(stopwords.words('english'))

# Mount Google Drive and load dataset
import pathlib
from google.colab import drive
drive.mount('/content/drive')
train_path = '/content/drive/My Drive/phm_train.csv'
test_path = '/content/drive/My Drive/phm_test.csv'
train_data = pd.read_csv(train_path)
test_data = pd.read_csv(test_path)

#Display a preview of datasets
print("Train Data Sample:")
print(train_data.head(), '\n')
print("\nTest Data Sample:")
print(test_data.head(), '\n')

# Preprocessing and Encoding labels
def preprocess_tweets(text_series):
    text_series = text_series.replace({'<.*?>': ''}, regex=True) # remove html tag
    text_series = text_series.replace({'[^A-Za-z]': ' '}, regex=True) # remove non alphabet
    text_series = text_series.apply(lambda tweet: [w.lower() for w in tweet.split() if w.lower() not in english_stops]) # lower and remove stopwords
    return text_series

x_train = preprocess_tweets(train_data['tweet'])
y_train = train_data['label'].astype(int)
x_test = preprocess_tweets(test_data['tweet'])
y_test = test_data['label'].astype(int)

print('Processed Training Set')
print(x_train.head(), '\n')
print('Processed Test Set')
print(x_test.head(), '\n')

# Function for getting the maximum tweet length, by calculating the mean of all the tweet lengths
def get_max_length(texts):
    tweet_length = []
    for tweet in texts:
        tweet_length.append(len(tweet))
    return int(np.ceil(np.mean(tweet_length)))

# Tokenize and Pad/Truncate Tweets
token = Tokenizer(lower=False)
token.fit_on_texts(x_train)
x_train_seq = token.texts_to_sequences(x_train)
x_test_seq = token.texts_to_sequences(x_test)
max_length = get_max_length(x_train)
x_train_pad = pad_sequences(x_train_seq, maxlen=max_length, padding='post', truncating='post')
x_test_pad = pad_sequences(x_test_seq, maxlen=max_length, padding='post', truncating='post')
total_words = len(token.word_index) + 1 # add 1 because of 0 padding

print('Encoded X Train\n', x_train_pad[:2], '\n')
print('Encoded X Test\n', x_test_pad[:2], '\n')
print('Maximum tweet length: ', max_length)

#-------------------------------------------------------------------------------
# LSTM model

EMBED_DIM = 32
LSTM_OUT = 64

model = Sequential()
model.add(Embedding(total_words, EMBED_DIM))
model.add(LSTM(LSTM_OUT))
model.add(Dense(1, activation='sigmoid'))

# Explicitly build the LSTM model before summary
model.build(input_shape=(None, max_length))
print(model.summary())

# Set hyperparameters
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

# Model Training
checkpoint = ModelCheckpoint(
    'models/LSTM.keras',
    monitor='val_accuracy',
    save_best_only=True,
    verbose=1
)

history_lstm = model.fit(x_train_pad, y_train, batch_size = 128, epochs = 3, validation_split=0.2, callbacks=[checkpoint])

# Model Testing
loss_lstm, acc_lstm = model.evaluate(x_test_pad, y_test, verbose=0)

pred_lstm = model.predict(x_test_pad)
y_pred_lstm = (pred_lstm >= 0.5)*1

true = 0
for i in range(len(y_pred_lstm)):
  if y_pred_lstm[i] == y_test[i]:
    true += 1


print('\nLSTM Test Details:')
print('Correct Predictions: {}'.format(true))
print('Wrong Predictions: {}'.format(len(y_pred_lstm) - true))
print('Test Accuracy: {:.2f}%'.format(acc_lstm*100))
print('Test Loss: {:.4f}\n'.format(loss_lstm))

#-------------------------------------------------------------------------------
# Bi-LSTM Model

model_bi = Sequential()
model_bi.add(Embedding(total_words, EMBED_DIM))
model_bi.add(Bidirectional(LSTM(LSTM_OUT)))
model_bi.add(Dense(1, activation='sigmoid'))

# Explicitly build the Bi-LSTM model before summary
model_bi.build(input_shape=(None, max_length))
print(model_bi.summary())

# Set hyperparameters
model_bi.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

# Model Training
checkpoint_bi = ModelCheckpoint(
    'models/BiLSTM.keras',
    monitor='val_accuracy',
    save_best_only=True,
    verbose=1
)

history_bi = model_bi.fit(x_train_pad, y_train, batch_size = 128, epochs = 3, validation_split=0.2, callbacks=[checkpoint_bi])

# Model Testing
loss_bilstm, acc_bilstm = model_bi.evaluate(x_test_pad, y_test, verbose=0)

pred_bilstm = model_bi.predict(x_test_pad)
y_pred_bilstm = (pred_bilstm >= 0.5)*1

true = 0
for i in range(len(y_pred_bilstm)):
  if y_pred_bilstm[i] == y_test[i]:
    true += 1


print('\nBi-LSTM Test Details:')
print('Correct Predictions: {}'.format(true))
print('Wrong Predictions: {}'.format(len(y_pred_bilstm) - true))
print('Test Accuracy: {:.2f}%'.format(acc_bilstm*100))
print('Test Loss: {:.4f}\n'.format(loss_bilstm))

#-------------------------------------------------------------------------------
# Plots for Accuracy and Loss

epochs = len(history_lstm.history['accuracy'])

plt.figure(figsize=(12,5))

# Accuracy plot
plt.subplot(1,2,1)
plt.plot(range(1, epochs+1), history_lstm.history['accuracy'], label='LSTM Train')
plt.plot(range(1, epochs+1), history_lstm.history['val_accuracy'], label='LSTM Val')
plt.plot(range(1, epochs+1), history_bi.history['accuracy'], label='Bi-LSTM Train')
plt.plot(range(1, epochs+1), history_bi.history['val_accuracy'], label='Bi-LSTM Val')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

# Loss plot
plt.subplot(1,2,2)
plt.plot(range(1, epochs+1), history_lstm.history['loss'], label='LSTM Train')
plt.plot(range(1, epochs+1), history_lstm.history['val_loss'], label='LSTM Val')
plt.plot(range(1, epochs+1), history_bi.history['loss'], label='Bi-LSTM Train')
plt.plot(range(1, epochs+1), history_bi.history['val_loss'], label='Bi-LSTM Val')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()

# Performance Comparison Table
print("\nPerformance Comparison Table:")
print(f"{'Model':<10} | {'Test Accuracy':<15} | {'Test Loss':<10}")
print(f"{'-'*40}")
print(f"{'LSTM':<10} | {acc_lstm*100:>13.2f}% | {loss_lstm:>10.4f}")
print(f"{'Bi-LSTM':<10} | {acc_bilstm*100:>13.2f}% | {loss_bilstm:>10.4f}")
```
